## Basic plotting using functional method

Using Numpy, two arrays are created. One which consists of equally spaced 11 elements from 1 to 5. The other is just the element multiplied by 2. Then the 'plot' function is called which takes in the two array as arguments and plots the graph.

```python
import numpy as np
a = np.linspace(0,5,11)
b = a ** 3
```

```python
a
```

```python
b
```

```python
plt.plot(a,b)
```

To add color and line style, mention the abreviation of the color followed by the linestyle type. For example if you want the color red and dotted line.

```python
plt.plot(a,b,'r--')
```

You can also add the lable and title for the plot. 

```python
plt.plot(a,b,'b-.')
plt.xlabel('a')
plt.ylabel('b')
plt.title('Graph')
```

## Creating multiplots using functional method

Multiple plots will be created using subplots. 'subplots' take into arguments the number of rows, the number of columns and the plot we are referring to. For example, here there is 1 row, 2 columns and we are referring to the first subplot.

```python
plt.subplot(1,2,1)
plt.plot(a,b,'g')

plt.subplot(1,2,2)
plt.plot(b,a,'y')
```

## Basic plotting using object oriented method

The basic idea of using this method is to create objects of the figures and then calling off methods on that object. So first we will create a figure object by writing the below command.

```python
fig  = plt.figure()
```

Now a figure object is created which is like a blank canvas. Now to do we will add the axes. The 'add_axes' method will take as an argument a list. The list will have 4 elements 'left', 'bottom', 'width' and 'height' which ranges from 0-1. Using the below exmaple, it means that the axes should be 10% in from left, 10% up from bottom, 80% of the canvas width and 80% of the canvas width.

```python
fig  = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
```

Now we will plot 'a' and 'b' on this axes.

```python
fig  = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(a,b)
axes.set_xlabel('a')
axes.set_ylabel('b')
axes.set_title('Graph')
```

The color of the plot can also be changed. The color can be mentioned directly or else the RBG hex code can also be mentioned. The linewidth can also be changed which takes in a integer value and changes the linewidth with respect to the default value. If the linewidth value is 10 then the width changes 10 times wrt to the default value. To work with transperacy you can mention the value of alpha. The linestyle is mentioned which can be any pattern.

```python
fig  = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(a,b,color='green',linewidth=10,alpha=0.5,linestyle='-.')
```

Now if the data points have to be shown then markers have to be mentioned. The markers can be o, *, + or even a number. Along with this the marker color and edge can also be customised.

```python
fig  = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(a,b,color='purple',linewidth=3,marker='o', markersize=10,
         markerfacecolor='red', markeredgewidth=2, markeredgecolor='yellow')
```

Suppose you want to only show the plot for a particular range of values, it can be done by setting the limits.

```python
fig  = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(a,b,color='black')

axes.set_xlim([0,2])
axes.set_ylim([0,3])
```

## Creating multiplots using object oriented method

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.3,0.4,0.5,0.3])
```

Now we will plot on both the axes.

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.15,0.5,0.3,0.3])

axes1.plot(a,b)
axes1.set_title('Bigger plot')

axes2.plot(b,a)
axes2.set_title('Smaller plot')
```

We can also use the 'subplots' method to create multiple plots. When using this, the 'subplots' method automatically decids the axes arguments unlike in the above example where we mentioned the arguments when using the 'figure' method. 

```python
fig,axes = plt.subplots(nrows=1, ncols=3)
```

To avoid the overlapping, use the 'tight_layout' method.

```python
fig,axes = plt.subplots(nrows=1, ncols=3)
plt.tight_layout()
```

Now plotting in the individual axes using indexing.

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()

axes[0].plot(a,b)
axes[0].set_title('Graph 1')

axes[1].plot(b,a)
axes[1].set_title('Graph 2')
```

## Adjusting the Figure size and DPI

Matplotlib provides the freedom to control these parameters while creating a figure object. The argument figsize takes in a tuple where the elements are width and height of the figure in inches. Another argument that can be passed is DPI. The full form of DPI is dots per inch. It takes in an integer value. It determines how many pixels the figure comprises. A larger dpi will act like a magnifying glass as shown below.

```python
fig = plt.figure(figsize=(4,3))

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(a,b)
```

```python
fig = plt.figure(figsize=(4,3), dpi=110)

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(a,b)
```

```python
fig = plt.figure(figsize=(9,3))

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(a,b)
```

To deal with figure size and DPI in subplots, the procedure is the same as shown below.

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(5,2))

axes[0].plot(a,b)
axes[1].plot(b,a)
```

## Dealing with legends

Legends is used to denote which plot is for what expression when there are multiple plots in a single figure.

```python
fig = plt.figure(figsize=(6,4))

axes = fig.add_axes([0,0,1,1])
axes.plot(a,b,label='b v/s a')
axes.plot(b,a,label='a v/s b')

axes.legend()
```

Incase the legend is overlapping the plot then you can define the location of the legend. The different locations that can be used are shown below. Incase none of the options work for you then you can also mention a tuple with the x and y coordinate.

```python
fig = plt.figure(figsize=(6,4))

axes = fig.add_axes([0,0,1,1])
axes.plot(a,b,label='b v/s a')
axes.plot(b,a,label='a v/s b')

axes.legend(loc=10)
```
