# Seaborn

## Importing Seaborn

Seaborn is imported and along with it the magic function is mentioned below it to make sure that the plots generated are inline and displayed in the notebook.

```python
import seaborn as sns
%matplotlib inline
```

## Load dataset

Seaborn has inbuilt datasets which can be used. So the 'tips' dataset will be used here. This dataset has information of how much tip is given on the total bill.

```python
dataset = sns.load_dataset('tips')
```

```python
dataset.head()
```

## Distribution plots

Distribution plots, also called as dist plot are created when there is a distribution of data which has to be visualised. It helps to visualize the distribution of univariate (one variable) set of data. When calling the 'distplot' method, the column of a dataframe is passed as an argument. The output is a hisogram with the KDE plot. The KDE can be removed by mentioning 'kde' = false. KDE stands for kernel density estimation. 

```python
sns.distplot(dataset['total_bill'])
```

```python
sns.distplot(dataset['total_bill'], kde=False)
```

To view the grid lines, then set the seaborn style to 'whitegrid'. The different style available are darkgrid, ticks, white. Also te bins can be changed by setting the 'bin' variable accoridng to the dataset.

```python
sns.set_style('whitegrid')
sns.distplot(dataset['total_bill'], kde=False, bins=40)
```

From the histogram we can draw the insight that most bills fall in the category of 10-20 dollars. 

## Joint plot

This type of plot helps to join two dist plots for bivariate (2 variables) data. The 'jointplot' method is called and then the x and y varibale is passed as arguments. These two varibales are basically the two columns we want to compare. Then the dataset is also passed

```python
sns.jointplot(x='total_bill', y='tip', data=dataset)
```

So on the x axis there is the histogram for the total bill and on the y axis there is a histogram for the tip. In the center is a scatter plot which shows a trend between the total bill and the tip. As the total bill is getting higher, the value of tip is also increasing. There is another argument called 'kind' whose default value is scatter and therefore a scatter plot is shown. It can be changed to a hex plot too. In a hex plot, the cells are darker if more data points are present.

```python
sns.jointplot(x='total_bill', y='tip', data=dataset, kind='hex')
```

Another kind is called regression. This displays a scatter plot but along with it draws a regression line too. This regression line just displays the linear relationship between the two varibales in form of a line.

```python
sns.jointplot(x='total_bill', y='tip', data=dataset, kind='reg')
```

## Pair plots

This type of plot graphs the pairwise relationships of the numerical columns across the entire dataframe. When calling the pairplot method, the dataframe is passed.

```python
sns.pairplot(dataset)
```

Incase there are any categorical columns i.e non-numerical columns then it can be passed in the argument 'hue'. It will show the plotting wrt this columns.

```python
sns.pairplot(dataset, hue='sex')
```

The color of the plot can also be changed by mentioning in the 'palette' argument.

```python
sns.pairplot(dataset, hue='sex', palette='rainbow')
```

## Rug plots

This plot displays a single mark along an axis and is used for univariate data. It is analogous to the histogram (dist plot) and is used to visualize distribution of the data. In histograms, bins are used for reperesntation and here the marks are used. The 'rugplot' method is called and a column is passed as an argument. The KDE plot is made using the rug plot. For each dah mark, a nomral distribution curve is formed and then all these curves are summed to form the KDE plot.

```python
sns.rugplot(dataset['total_bill'])
```

## Categorical plots

Categorical means the non-numerical values. Up until now, we have seen the ways to plot using numerical data, now we will see how we can deal with the categorical data plotting and view the distribution of the categorical columns.

## Bar plots

This type of categorical plot gives the provision to aggregate the categorical data based on a function like mean. It is analogous to the 'group by' method. The 'barplot' method takes as argument x varibale, y variable and the dataframe. So the x variable can be a categorical columns and the y varibale can be a numerical column.

```python
sns.barplot(x='sex',y='total_bill',data=dataset)
```

The plot above shows the mean total bill of males and females since the default estimator is mean. But this estimator can be changed by passing another value to the argument 'estimator'. The estimator could be variance or standard deviation also.

```python
import numpy as np

sns.barplot(x='sex',y='total_bill',data=dataset,estimator=np.std)
```

## Count plot

This type of plot is just like the bar plot but here the estimator is the count or the number of occurences. Here the method 'countplot' is called and only the x variable along with the dataframe is passed. The counting is done for this x variable. 

```python
sns.countplot(x='sex', data=dataset)
```

Incase you want to remove the spines surrounding the plot then call the 'despine' method. It removes the top and right spine. Incase you want to remove the left and bottom spine then pass True value for the arguments left and bottom.

```python
sns.countplot(x='sex', data=dataset)
sns.despine()
```

## Box plots

Box plots are also called as whisker plots. It shows the distribution of the categorical data. The method 'boxplot' takes as argument x and variable along with the dataframe name. The x variable is the categorical column and y variable is the numerical column. 

```python
sns.boxplot(x='day',y='total_bill',data=dataset)
```

The box plot displays the quartiles of the dataset along with the whiskers which extend to show the rest of the dirtribution. The points which lie outside of the whiskers are called as outliers. The lower horizontal line represents the minimum value and the higher vertical line represents the maximum value. The middle line in the box denotes the median. The first quartile (25%) is the middle number between the smallest number (not the “minimum”) and the median of the dataset. The third quartile (75%) is the middle value between the median and the highest value (not the “maximum”) of the dataset.

Another arguments called 'hue' can be passed which further represnts the box plots according to this.

```python
sns.boxplot(x='day',y='total_bill',data=dataset,hue='sex')
```

## Violin plots

This type of plot also represents the distribution of the categorical columns. It takes the same arguments like the box plot. It is a considered a combination of the box plot with a kernel density plot. 

```python
sns.violinplot(x='day',y='total_bill',data=dataset)
```

The violin plot also has the provison to provide the hue. But there is an extra functionality of splitting the distribution according to the hue and displaying together.

```python
sns.violinplot(x='day',y='total_bill',data=dataset,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=dataset,hue='sex',split=True)
```

## Strip plot

This kind of plot displays the scatter plot where one value is categorical and other is numerical. It takes the arguments x and y variable and the dataframe name.

```python
sns.stripplot(x='day',y='total_bill',data=dataset)
```

The hue can be added here too along with the split argument and it will plot accordingly.

```python
sns.stripplot(x='day',y='total_bill',data=dataset,hue='smoker',split=True)
```

## Swarm plot

This type of plot is a combination of the strip plot and the violin plot. It takes in the x and y variable along with the dataframe name.

```python
sns.swarmplot(x='day',y='total_bill',data=dataset)
```

## Factor plot

It take the x and y variable along with the dataframe name and hue. But what makes it different is that it takes a 'kind' argument too. By mentioning the kind value, any plot can be generated.

```python
sns.factorplot(x='day',y='total_bill',data=dataset,kind='bar')
```

```python
sns.factorplot(x='day',y='total_bill',data=dataset,kind='violin')
```

## Heatmaps

Heatmaps are also called as matrix plots. To be able to plot this the data should be in the form of a matrix. The data is in matrix form if there is a row name and the column name. 

```python
dataset.head()
```

Here for the first cell, the column name is 'total_bill' and the row value is 0. So it is not in the matrix form. To convert into matrix form, we can find the correlation.

```python
dataset.corr()
```

Now according to the column and row name, we can get the relevancy value. Based on these values the heatmap is drawn. The heatmap is a colored representation of these values obtained.

```python
sns.heatmap(dataset.corr())
```

If you want to display the values, then pass the argument 'annot' as True. The color of the heatmap can also be changed by passing the desired value in the 'cmap' argument.

```python
sns.heatmap(dataset.corr(),annot=True,cmap='magma')
```

The correlation can also be obtained using the pivot table concept. To understand this better, consider another inbuilt seaborn dataset 'flights'.

```python
dataset2 = sns.load_dataset('flights')
```

```python
dataset2.head()
```

Now using pivot table, the month will become the row and years will become the column along with the passengers as values. This way the column and the row will be named and will be in the matrix form. Then the heatmap can be plotted. Incase the cells are touching each other too much in the heatmap, the linewidth and linecolor can also be mentioned to show distinct cells.

```python
df = dataset2.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(df,cmap='coolwarm',linecolor='black',linewidth=1)
```

## Cluster map

This map is produced using the concept of hierarchial clustering to produce the clustered version of the heatmap. The simlar columns and rows are clustred together.

```python
sns.clustermap(df)
```

## Grids

To understand how grids work, we will use the very famous flower dataset called iris dataset. 

```python
import matplotlib.pyplot as plt
dataset3 = sns.load_dataset('iris')
dataset3.head()
```

Using seaborn's pair grid mechanism, we will customise the pairplots. We call the 'PairGrid' method and pass the dataframe as argument. It creates empty grids and this gives us more control over what to display. 

```python
grid = sns.PairGrid(dataset3)
```

Now if we want a scatter plot, then we can map it to the grid.

```python
grid.map(plt.scatter)
```

We can also specify what we need in the diagonals, upper part and lowe part.

```python
grid1 = sns.PairGrid(dataset3)
grid1.map_diag(sns.distplot)
grid1.map_upper(plt.scatter)
grid1.map_lower(sns.kdeplot)
```

## Facet grid

To demonstrate facet grid, the tips dataset will be used. The FacetGrid method is called and the dataframe is passed. Along with this the column and the row value is also passed.

```python
grid2 = sns.FacetGrid(data=dataset,col='time',row='smoker' )
```

This forms an empty grid. Now we can plot in these grids. So we can plot a dist plot on total bills. The result we get is according to the rows and columns mentioned and they are separted on the basis of it. So as seen below, the left column is lunch and right column is dinner because column value is mentioned s time. Then the upper row is smoker-yes and lower row is smoker-no since the row value is of smoker. The x-axis is the total bill. 

```python
grid2.map(sns.distplot,'total_bill')
```

We can pass more columns too where it is needed. Like in scatter plot we need two variables.

```python
grid2.map(plt.scatter,'total_bill','tip')
```

## Regression plot

The regression plot helps to depict the linear regression models. We will look into the LM plot or line plot. It takes into arguments the x and y varibale along with the dataset and plots the best fit line.

```python
sns.lmplot(x='total_bill',y='tip',data=dataset)
```

The 'hue' argument can be mentioned and according to the hue the markers can also be changed. 

```python
sns.lmplot(x='total_bill',y='tip',data=dataset,hue='sex',markers=['o','+'])
```

You can separate it according to the column or rows instead of using the 'hue' argument to separate.

```python
sns.lmplot(x='total_bill',y='tip',data=dataset,col='sex')
```
